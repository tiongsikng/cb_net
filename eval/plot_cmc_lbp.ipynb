```python
import matplotlib.pyplot as plt
import torch
import seaborn as sns
import numpy as np

plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
plt.rcParams.update({'font.size': 17, 'legend.fontsize': 17})    

rank = 10
rng = np.arange(0, rank) + 1
ticks = np.arange(0, (len(rng) + 1), 1)
```

```python
peri_cmc_dict = torch.load('../data/cmc/lbp/peri/peri_cmc_dict_lbp.pt')        
peri_avg_dict = torch.load('../data/cmc/lbp/peri/peri_avg_dict_lbp.pt')
face_cmc_dict = torch.load('../data/cmc/lbp/face/face_cmc_dict_lbp.pt')
face_avg_dict = torch.load('../data/cmc/lbp/face/face_avg_dict_lbp.pt')        
```

```python
print(peri_cmc_dict)
print(face_cmc_dict)
print(peri_avg_dict)
print(face_avg_dict)
```

```python
cm_base_cmc_dict_f = torch.load('../data/cmc/lbp/cm/cm_base_cmc_dict_f_lbp.pt')
cm_base_avg_dict_f = torch.load('../data/cmc/lbp/cm/cm_base_avg_dict_f_lbp.pt')
cm_base_cmc_dict_p = torch.load('../data/cmc/lbp/cm/cm_base_cmc_dict_p_lbp.pt')
cm_base_avg_dict_p = torch.load('../data/cmc/lbp/cm/cm_base_avg_dict_p_lbp.pt')

cm_base_avg_dict = (cm_base_avg_dict_f['avg'] + cm_base_avg_dict_p['avg']) / 2
```

```python
print(cm_base_cmc_dict_f)
print(cm_base_cmc_dict_p)
```

```python
mm_base_cmc_dict_score = torch.load('../data/cmc/lbp/mm/mm_cmc_dict_score_lbp.pt')        
mm_base_avg_dict_score = torch.load('../data/cmc/lbp/mm/mm_cmc_avg_dict_score_lbp.pt')
```

```python
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
plt.rcParams.update({'font.size': 17, 'legend.fontsize': 17})    
fig = plt.figure()
fig.set_figheight(5.5)
ax = fig.add_subplot(1, 1, 1)
plt.grid(which='both')
plt.xticks(ticks)   
    
ax.plot(rng, 100 * peri_avg_dict['avg'], label='LBP (Periocular)', marker = 'x', color = '#d80073', linestyle='--')
ax.plot(rng, 100 * face_avg_dict['avg'], label='LBP (Face)', marker = 'x', color = '#1ba1e2', linestyle='--')
ax.plot(rng, 100 * cm_base_avg_dict, label='LBP (Cross-Modal)', marker = 'x', color = 'darkviolet', linestyle='--')
ax.plot(rng, 100 * mm_base_avg_dict_score['avg'], label='LBP (Multiple Modalities)', marker = 'x', color = 'darkgreen', linestyle='--')
plt.xlabel('Rank')
plt.ylabel('Identification Rate (%)')
plt.legend(bbox_to_anchor=(1, 1), loc="upper left")
plt.savefig('../graphs/cmc/lbp_CMC.pdf', bbox_inches='tight')
plt.show()
plt.clf()
```
